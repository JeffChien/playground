```python
def df_info2(df):
    df_info = pd.DataFrame()
    df_info['dtype'] = df.dtypes
    df_info['#'] = df.count()
    df_info['#na'] = df.isna().sum()
    df_info['#unique'] = df.nunique()
    df_info['min'] = df.min()
    df_info['max'] = df.max()
    return df_info
```

```python
df_info2(df)
```
